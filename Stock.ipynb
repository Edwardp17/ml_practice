```python
import pandas as pd
import os
```

```python
header = ['url']
file = pd.read_csv('urls.txt',names=header)
```

```python
urls = []
for index, row in file.iterrows():
    urls.append(row['url'])
```

```python
frames = {}
headers = ['date','day','open','high','low','close','volume']
i = 0
for url in urls:
    i = i+1
    df = pd.read_csv(url,names=headers)
    df = df.drop(df.columns[1:6],1)
    frames[i]=df
```

```python
frameoutput = pd.DataFrame()
for frame in frames:
    frameoutput = pd.concat([frameoutput,frames[frame]],1)
```
